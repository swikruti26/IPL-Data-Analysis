# Exploring the Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("IPL Data.xlsx")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
f,ax = plt.subplots(figsize=(14,10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", ax=ax)
plt.show()
```

Heatmap was used to check for correlation between various columns of the dataset. 

# Number of Balls in the Tournament

```python
balls = df.ball
number_of_balls = len(balls)
print("Number of Balls in the Tournament: " + str(number_of_balls))
```

As the balls have values of 1,2,3,4,5 and 6; adding these values would not be correct. Hence, length of the column was taken as the number of balls. 

# Number of Wickets in the Tournament

```python
wickets = df.dismissal_kind.count()
print("Number of Wickets in the Tournament: " + str(wickets))
```

To calculate the number of wickets, non-null values of the "dismissal_kind" column were counted. 

# Q1. Balls per Wickets

```python
#Balls per wicket

balls_per_wicket = number_of_balls//wickets
print("Balls per wicket is: " + str(balls_per_wicket))
```

# Number of Sixes in the Tournament

```python
sixes = 0
for elements in df.batsman_runs:
  if elements == 6:
    sixes = sixes + 1

print("Number of sixes is: "+str(sixes))
```

# Q1. Balls per Sixes

```python
#Balls per sixes

balls_per_sixes = number_of_balls//sixes 
print("Balls per sixes is: "+str(balls_per_sixes))
```

# Total Runs in the Tournament

```python
runs = df.batsman_runs
extras = df.extra_runs

total_runs = runs.sum() + extras.sum()
print("Total Runs in the Tournament: " + str(total_runs))
```

# Q1. Runs per Wicket

```python
#Runs per wicket

runs_per_wicket = total_runs//wickets
print("Runs per wicket is: " + str(runs_per_wicket))
```

# Batsman Data

# Total Runs by each Batsman

```python
total_runs_batsmen = df.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
total_runs_batsmen
```

# Total Dots by each Batsman

```python
dots = df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
dots
```

# Balls Faced by each Batsman

```python
ball_faced = df.groupby('batsman')['batsman_runs'].agg('count').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
ball_faced
```

# Final Batting Stats

```python
f = pd.merge(total_runs_batsmen, dots, on='batsman')
final = pd.merge(f,ball_faced, on='batsman')
final.rename(columns={'batsman_runs_x': 'Total Runs', 'batsman_runs_y': 'Dots', 'batsman_runs': 'Balls'}, inplace=True)
final
```

# Q2. The ratio of total balls to total dots for each batsman

```python
#The ratio of total balls to total dots for each batsman
final['Balls to Dots'] = final.Balls/final.Dots
final
```

# Sixes by each Batsman

```python
sixes_by_batsman = df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
sixes_by_batsman
```

# Fours by each Batsman

```python
fours_by_batsman = df.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
fours_by_batsman
```

# Total Boundaries (4s and 6s) by each Batsman

```python
boundaries = pd.merge(sixes_by_batsman,fours_by_batsman, on='batsman')
boundaries.rename(columns={'batsman_runs_x': 'Sixes', 'batsman_runs_y': 'Fours'}, inplace=True)
boundaries['Boundaries'] = boundaries.Sixes+boundaries.Fours
boundaries
```

```python
new_dfs = [total_runs_batsmen,boundaries]
boundary_to_runs = pd.merge(total_runs_batsmen,boundaries, on='batsman')
boundary_to_runs
```

# Runs by boundaries for each Batsman

```python
boundary_to_runs['Total'] = 6*boundary_to_runs.Sixes + 4*boundary_to_runs.Fours 
boundary_to_runs 
```

# Q3. Percentage of boundary runs to total runs for each batsman

```python
#Percentage of boundary runs to total runs for each batsman
boundary_to_runs['Percentage of Boundary to Total Runs'] = (boundary_to_runs.Total/boundary_to_runs.batsman_runs)*100
boundary_to_runs
```

# Top 20 Batsmen with the highest percentage of boundary to total runs

```python
higest_boundary_percent = boundary_to_runs.iloc[:20,:]
higest_boundary_percent = higest_boundary_percent.sort_values(by='Percentage of Boundary to Total Runs',ascending=False).reset_index(drop=True)
plt.figure(figsize=(25,10))
plt.bar(higest_boundary_percent.batsman, higest_boundary_percent['Percentage of Boundary to Total Runs'])
plt.xlabel('Batsmen')
plt.ylabel("Boundary Percentage")
plt.title("Batsmen with the highest percentage of boundary to total runs in IPL")

plt.show()
```

# Q5. Best Death Over Bowler

# Top Scores by Batsmen in IPL
